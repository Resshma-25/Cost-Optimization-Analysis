```python
import pandas as pd

df = pd.read_csv("Cost_data.csv")
df
```

```python
df.info()
df.describe()
```

The dataset contains product-level cost information across multiple departments. It includes hardware-related items with varying cost ranges. Initial analysis shows significant cost variation across departments, making it suitable for cost optimization analysis.


```python
dept_cost = df.groupby("Department")["Cost"].sum().reset_index()
dept_cost
```

The IT department contributes the highest total cost compared to other departments. This indicates that cost optimization initiatives should prioritize IT-related hardware and procurement strategies.

```python
high_cost_products = df[df["Cost"] > df["Cost"].mean()]
high_cost_products
```

Products with costs above the overall average are potential optimization candidates. These items may benefit from supplier renegotiation, alternative sourcing, or design optimization.

```python
import matplotlib.pyplot as plt

plt.bar(dept_cost["Department"], dept_cost["Cost"])
plt.title("Department-wise Cost Distribution")
plt.xlabel("Department")
plt.ylabel("Total Cost")
plt.show()
```

Visualization clearly highlights cost concentration across departments, helping stakeholders quickly identify priority areas.

Business Recommendation

Prioritize IT department for cost optimization initiatives.

Review procurement strategies for high-cost hardware products.

Perform supplier benchmarking to identify cost-saving opportunities.

Explore alternative components without impacting system performance.
